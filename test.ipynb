```python
import PyPDF2

# Open PDF file
pdf_path = "invoice_vet.pdf"

with open("invoice_vet.pdf", "rb") as file:
    reader = PyPDF2.PdfReader(file)
    print("Number of pages:", len(reader.pages))
    
    # Extract text from each page
    for i, page in enumerate(reader.pages):
        text = page.extract_text()
        print(f"\n--- Page {i+1} ---\n")
        print(text)
```

```python
import pdfplumber
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        tables = page.extract_tables()
        if tables:
            print(f"\n✅ Table(s) found on Page {i+1}:")
            print(tables)
            print("-----------------------------")
            # for table in tables:
            #     for row in table:
            #         print(row)
        else:
            print(f"\nNo table found on Page {i+1}")
```

```python
all_tables_dicts = []  # store all tables as list of dicts

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        tables = page.extract_tables()
        if tables:
            print(f"\n✅ Table(s) found on Page {i+1}:")
            for table in tables:
                headers = table[0]   # first row is header
                rows = table[1:]     # rest are data
                dict_rows = [dict(zip(headers, row)) for row in rows]
                all_tables_dicts.append(dict_rows)
                print(all_tables_dicts)

                # print nicely
                # for row in dict_rows:
                #     print(row)
                print("-----------------------------")
        else:
            print(f"\nNo table found on Page {i+1}")
```
